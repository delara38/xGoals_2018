# Exploring different ways to visualize xGoals Through Matplotlib 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from mpl_toolkits.mplot3d import Axes3D
import math
from matplotlib import cm
import matplotlib.image as mpimg
```

```python
model = pickle.load(open('xGoals_model.sav','rb'))
```

```python
@np.vectorize
def func(x,y):
    dist = np.sqrt( x** 2 + y**2)
    angle = math.atan2(y,x) * 180 / math.pi
    if angle == 90:
        angle = 0
    elif angle < 90:
        angle = 90- angle
    elif angle > 90:
        angle = angle - 90
    
    predict = model.predict_proba(np.array([dist,angle]).reshape(1,2)).tolist()[0][1]
    return predict

@np.vectorize
def func1(x,y):
    y = y * (64/50)
    dist = np.sqrt( x** 2 + y**2)
    angle = math.atan2(y,x) * 180 / math.pi
    if angle == 90:
        angle = 0
    elif angle < 90:
        angle = 90- angle
    elif angle > 90:
        angle = angle - 90
    
    predict = model.predict_proba(np.array([dist,angle]).reshape(1,2)).tolist()[0][1]
    return np.log(predict)
    
```

# 3D Surface Plots
here we plot a 3d graph with the z axis being xGoals and then x and y being on ice co-ordinates. Advantages of this are that we can easily compare two different spots. Disadvantages are that it is hard to understand if you dont already know what the graph is meant to do because it is not plotted on a rink and relies on imagination for putting it over an offensive zone

```python
y = np.linspace(0,64,64)
x = np.linspace(-42,42,84)
X,Y = np.meshgrid(x,y)
X = X.T
Y = Y.T

Z = func(X,Y)
```

```python
fig = plt.figure(figsize= [30,10])
ax = fig.add_subplot(131, projection='3d')
ax.plot_surface(X,Y,Z, cmap= cm.coolwarm)
ax.view_init(10,90)


ax = fig.add_subplot(132, projection='3d')
ax.plot_surface(X,Y,Z, cmap= cm.coolwarm)
ax.view_init(20, -10)

ax = fig.add_subplot(133, projection='3d')
ax.plot_surface(X,Y,Z, cmap= cm.coolwarm)
ax.view_init(30, 50)

plt.show()
```

# Contour Plots
Here we make a contour plot where the more xG an area has lines will curve around it more. advantages of this are its much easier to understand but you cant compare different shooting spots as well as you can on a surface plot. these do not look very nice but if you go to hockeyviz.com you can see very nice contour graphs using shotx instead of xG

```python
img = plt.imread('rink.png')
fig, ax = plt.subplots()
ax.imshow(img, extent = [-42,42,50,-11])
```

```python
y = np.linspace(0,50,64)
x = np.linspace(-40,40,80)
X,Y = np.meshgrid(x,y)
X = X.T
Y = Y.T

Z = func1(X,Y)
```

```python
img = plt.imread('rink.png')
#fig, ax = plt.subplots(figsize=[8,8])
plt.imshow(img, extent = [-42,42,50,-11])
plt.contourf(X,Y,Z, levels = 8,cmap = cm.coolwarm, alpha = 0.9)
plt.colorbar()
```

# Plotting xG onto the ice by x and y Co-ordinates
This method is good for an understanding of what xGoals does in general but, it is not 100% reflective of how our model decides xGoals

```python
x = []
y = []
z = []
for f in range(0,25):

    f = f*2

    for g in range(0,41):
        g = (g*2) - 40
        if not((f==0) and (g == -40 or g == 40)):
            x.append(g)
            
            y.append(f)
            f = f
            
            dist = np.sqrt(g**2 + f**2)
            angle = math.atan2(f,g) * 180 / math.pi
            
            if angle == 90:
                angle = 0
            elif angle < 90:
                angle = 90 - angle
            elif angle > 90:
                angle = angle - 90
            
            pred = model.predict_proba(np.array([dist,angle]).reshape(1,2))
            z.append(np.log(pred.tolist()[0][1]))
```

```python
plt.imshow(img, extent = [-42,42,50,-11])
plt.scatter(x,y,c = z,alpha=0.75, cmap = cm.coolwarm)
plt.plot()
```
